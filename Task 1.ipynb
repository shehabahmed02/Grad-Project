```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


```

```python
# Summary Statistics
summary_stats = data.describe()
print("Summary Statistics:")
print(summary_stats)
```

```python
# Data Visualization - Histograms for numerical columns
numerical_cols = data.select_dtypes(include=np.number).columns.tolist()

for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    plt.hist(data[col], bins=20, edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Data Visualization - Bar plots for categorical columns
categorical_cols = data.select_dtypes(include='object').columns.tolist()

for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    sns.countplot(data[col])
    plt.title(f'Bar plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Correlation Analysis
correlation_matrix = data.corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Visualization of Correlation Matrix using heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Value Counts for Categorical Columns
for col in categorical_cols:
    value_counts = data[col].value_counts()
    print(f"Value Counts for {col}:")
    print(value_counts)
```

```python
# Filter the dataset for the specified match
match_info = data[(data['Home'] == 'Brentford') & (data['Away'] == 'Manchester City') & (data['Season_End_Year'] == 2023) & (data['Wk'] == 38)]

# Retrieve the Full Time Result (FTR) of the match
result = match_info['FTR'].values[0] if not match_info.empty else "Match not found"

print(f"The result of Brentford vs Manchester City match in 2023, game week 38 was: {result}")
```

```python
# Filter the dataset for the specified match
match_info = data[(data['Home'] == 'Brentford') & (data['Away'] == 'Manchester City') & (data['Season_End_Year'] == 2023) & (data['Wk'] == 38)]

# Retrieve the HomeGoals and AwayGoals of the match
if not match_info.empty:
    home_goals = match_info['HomeGoals'].values[0]
    away_goals = match_info['AwayGoals'].values[0]
    print(f"The result of Brentford vs Manchester City match in 2023, game week 38 was - Brentford {home_goals} : {away_goals} Manchester City")
else:
    print("Match not found")
```

```python
# Filter the dataset for matches between Manchester City and Manchester United from 2000 to 2023
manchester_matches = data[
    (
        ((data['Home'].str.contains('Manchester City')) | (data['Away'].str.contains('Manchester City'))) &
        ((data['Home'].str.contains('Manchester United')) | (data['Away'].str.contains('Manchester United')))
    ) & (data['Season_End_Year'] >= 2000) & (data['Season_End_Year'] <= 2023)
]

# Filter for Manchester City home games against Manchester United
man_city_home = manchester_matches[
    (manchester_matches['Home'].str.contains('Manchester City')) & (manchester_matches['Away'].str.contains('Manchester United'))
]

# Filter for Manchester City away games against Manchester United
man_city_away = manchester_matches[
    (manchester_matches['Away'].str.contains('Manchester City')) & (manchester_matches['Home'].str.contains('Manchester United'))
]

# Combine the filtered dataframes to get all Manchester City vs Manchester United matches
all_manchester_matches = pd.concat([man_city_home, man_city_away])

# Displaying the relevant columns for home goals and away goals
selected_columns = ['Season_End_Year', 'Home', 'HomeGoals', 'Away', 'AwayGoals']
goals_info = all_manchester_matches[selected_columns]

print(goals_info)
```
